Importing Libraries

```python
import pandas as pd
import numpy as np
import joblib

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier
```

Loading Train Set

```python
train = pd.read_csv(r'C:\Users\pc\Documents\MACHINE LEARNING\Codveda-Internship\KNN Classifier\datasets\iris train set.csv')
train.head()
```

Feature Selection

```python
x = train.drop('species', axis=1)
y = train['species']
```

Train Set Splitting 

```python
x_train, x_validate, y_train, y_validate = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)
```

Model Development

KNN when K=5 

```python
pipe1 = Pipeline(
    [
        ('scaler', StandardScaler()),
        ('knn', KNeighborsClassifier(n_neighbors=5))
    ]
)
```

```python
pipe1.fit(x_train, y_train)
```

```python
y_knn5_pred = pipe1.predict(x_validate)
```

Accuracy Score

```python
ac5 = accuracy_score(y_validate, y_knn5_pred)
print(f'Accuracy Score when KNN=7 is: {ac5}')
```

Confusion Matrix

```python
confusion_matrix(y_validate, y_knn5_pred)
```

Classification Report

```python
CR_k5 = classification_report(y_validate, y_knn5_pred, output_dict=True)
CR_k5 = pd.DataFrame(CR_k5)
CR_k5
```

KNN when K=3

```python
pipe2 = Pipeline(
    [
        ('scaler', StandardScaler()),
        ('knn', KNeighborsClassifier(n_neighbors=3))
    ]
)
```

```python
pipe2.fit(x_train, y_train)
```

```python
y_knn3_pred = pipe2.predict(x_validate)
```

Accuracy Score

```python
ac3 = accuracy_score(y_validate, y_knn3_pred)
print(f'Accuracy Score when KNN=7 is: {ac3}')
```

Confusion Matrix

```python
confusion_matrix(y_validate, y_knn3_pred)
```

Classification Report

```python
CR_k3 = classification_report(y_validate, y_knn3_pred, output_dict=True)
CR_k3 = pd.DataFrame(CR_k3)
CR_k3
```

KNN when K=7

```python
pipe3 = Pipeline(
    [
        ('scaler', StandardScaler()),
        ('knn', KNeighborsClassifier(n_neighbors=7))
    ]
)
```

```python
pipe3.fit(x_train, y_train)
```

```python
y_knn7_pred = pipe3.predict(x_validate)
```

Accuracy Score

```python
ac7 = accuracy_score(y_validate, y_knn7_pred)
print(f'Accuracy Score when KNN=7 is: {ac7}')
```

Confusion Matrix

```python
confusion_matrix(y_validate, y_knn7_pred)
```

Classification Report

```python
CR_k7 = classification_report(y_validate, y_knn7_pred, output_dict=True)
CR_k7 = pd.DataFrame(CR_k7)
CR_k7
```
